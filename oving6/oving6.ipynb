# Which features are most discriminitave?

```python
import pandas as pd
import numpy as np
import pandas.plotting
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df = pd.read_csv('agaricus-lepiota.data')
df.head()
```

```python
X = pd.get_dummies(df.drop('edibility',1))
y = pd.get_dummies(df.edibility)
X.head()
```

```python
skb = SelectKBest(chi2, k=10)
skb.fit(X, y)
X_new = skb.transform(X)

bestFeatures = np.array(X.columns)[skb.get_support(indices=True)]
```

```python
print("Best 10 Features:")
for feature in bestFeatures:
    print("- " + feature)
    
```

# Which combination of features explain the most variance in the dataset?

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
amount = 5
pca = PCA(n_components= amount)
pca.fit(X)
pcaframe = pd.DataFrame(pca.components_, columns=X.columns)

for num in range(amount):
    print(num+1)
    print(pcaframe.T[num].nlargest(10))
    print()
```
